# Domain Study

## Business Context:
#### No-Churn Telecom is a telecom company in Europe that has been around for over ten years. Recently, new companies have entered the market, making it harder for No-Churn to keep their customers from switching to competitors.
####
## Problem:
#### No-Churn Telecom has been trying to keep its customers happy by reducing prices and offering more deals. However, many customers (more than 10%) are still switching to other telecom companies. This high number of customers leaving is a big concern for No-Churn Telecom.
####
## Project Goals:
### 1) Understanding why Customers Leave:
#### Identify the key factors that make customers decide to leave No-Churn. By knowing these reasons, the company can address the issues and improve customer satisfaction.
### 2) Creating Churn Risk Scores:
#### Develop a model that assigns a "churn risk score" to each customer. This score indicates how likely a customer is to leave. Customers with higher scores are at greater risk of leaving.
### 3) Introducing CHURN-FLAG:
#### Create a new column called "CHURN-FLAG" in the customer data, with values YES (1) or NO (0). This flag will indicate whether a customer is likely to leave (YES) or stay (NO). Customers with a YES flag can be targeted with special offers to encourage them to stay.
####
## Features Information for understanding
#### 1) State: The state in which the customer resides. This is a categorical variable.
#### 2) Account Length: The duration (in days or months) that the customer has been with the company.
#### 3) Area Code: The area code of the customer's phone number.
#### 4) Phone: The customer's phone number.
#### 5) International Plan: Indicates whether the customer has an international plan (Yes/No).
#### 6) VMail Plan: Indicates whether the customer has a voice mail plan (Yes/No).
#### 7) VMail Message: The number of voice mail messages the customer has.
#### 8) Day Mins: The total number of minutes the customer spent on daytime calls.
#### 9) Day Calls: The total number of daytime calls made by the customer.
#### 10) Day Charge: The total charges for daytime calls.
#### 11) Eve Mins: The total number of minutes the customer spent on evening calls.
#### 12) Eve Calls: The total number of evening calls made by the customer.
#### Eve Charge: The total charges for evening calls.
#### Night Mins: The total number of minutes the customer spent on nighttime calls.
#### Night Calls: The total number of nighttime calls made by the customer.
#### Night Charge: The total charges for nighttime calls.
#### International Mins: The total number of minutes the customer spent on international calls.
#### International calls: The total number of international calls made by the customer.
#### International Charge: The total charges for international calls.
#### CustServ Calls: The number of calls made by the customer to customer service.
#### Churn: Indicates whether the customer has churned (Yes/No).

```python
#(Considering Time Period as one Variable and Avgrage Minutes as second.)
# Calculate the average minutes for each time period
avg_day_mins = df['Day Mins'].mean()
avg_eve_mins = df['Eve Mins'].mean()
avg_night_mins = df['Night Mins'].mean()

# Create a DataFrame for plotting
avg_mins_df = pd.DataFrame({
    'Time Period': ['Day', 'Evening', 'Night'],
    'Average Minutes': [avg_day_mins, avg_eve_mins, avg_night_mins]
})

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Time Period', y='Average Minutes', data=avg_mins_df, hue='Time Period', palette='viridis')

plt.title('Average Minutes at Different Times of the Day')
plt.xlabel('Time Period')
plt.ylabel('Average Minutes')
plt.show()
```

#### Observations show that customers spend more time on calls during the evening and night than during the day. To attract more customers, it could be effective to offer discounts on recharge plans specifically for evening and night calls.
####

```python
# Calculate the charge per minute for each time period
df['Day Charge per Minute'] = df['Day Charge'] / df['Day Mins']
df['Eve Charge per Minute'] = df['Eve Charge'] / df['Eve Mins']
df['Night Charge per Minute'] = df['Night Charge'] / df['Night Mins']

# Calculate the mean charges per minute
avg_day_charge_per_min = df['Day Charge per Minute'].mean()
avg_eve_charge_per_min = df['Eve Charge per Minute'].mean()
avg_night_charge_per_min = df['Night Charge per Minute'].mean()
grand_avg_charge_per_min = df[['Day Charge', 'Eve Charge', 'Night Charge']].sum().sum() / df[['Day Mins', 'Eve Mins', 'Night Mins']].sum().sum()

# Create a DataFrame for plotting
avg_charge_per_min_df = pd.DataFrame({
    'Time Period': ['Day', 'Evening', 'Night', 'Grand Average'],
    'Charge per Minute': [avg_day_charge_per_min, avg_eve_charge_per_min, avg_night_charge_per_min, grand_avg_charge_per_min]
})

# Plotting the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Time Period', y='Charge per Minute', data=avg_charge_per_min_df, hue='Time Period', palette='viridis')

plt.title('Charge per Minute for Different Times of the Day')
plt.xlabel('Time Period')
plt.ylabel('Charge per Minute')
plt.show()
```

#### It has been observed that the average charge per minute for calls made during the evening and night is lower than the grand average charge per minute. This is beneficial for attracting customers who prefer to make calls during these times. However, the average charge per minute for daytime calls is significantly higher than the grand average charge per minute. This could be a factor that some customers dislike about the company, potentially leading them to churn. Therefore, reducing the average charge per minute for daytime calls to at least match the grand average charge per minute could be an effective strategy to retain customers.

### Note: 
#### It is observed that there is a significant difference between the average charge per minute during different time periods. This difference may be due to chance or the presence of outliers. Therefore, it is beneficial to confirm this observation with hypothesis testing.

# Hypothesis Testing

#### Here, we have to compare the means of three different groups, so we use ANOVA. Before using it, we have to confirm that the assumptions of ANOVA are satisfied.

#### 1) Independence: The observations are independent. So first assumption is satisfied.
####

#### 2) Normality: The data within the group should be approximately Normal.

```python
# Creating New Columns.
# Calculate the charge per minute for each time period
df['Day Charge per Minute'] = df['Day Charge'] / df['Day Mins']
df['Eve Charge per Minute'] = df['Eve Charge'] / df['Eve Mins']
df['Night Charge per Minute'] = df['Night Charge'] / df['Night Mins']

# Verify the new columns
df[['Day Charge per Minute', 'Eve Charge per Minute', 'Night Charge per Minute']].head()
```

```python
features = ['Day Charge per Minute', 'Eve Charge per Minute', 'Night Charge per Minute']
data = df
plt.figure(figsize=(15,10))
i = 1
for feature in features:
    plt.subplot(2, 2, i)
    sns.kdeplot(data[feature], color = 'red')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')
    i = i+1
plt.tight_layout()
plt.show()
```

#### It is observed that the distribution of all the features is approximately normal, so the assumption of normality is satisfied.
####

#### 3) Homogeneity of Variances

```python
# Calculate variances
variances = df[features].var()

print("Variances of the specified columns:")
print(variances)
```

#### The variances of three groups are nearly equal so third assumption is also satisfied.

## ANOVA

### Hypotheses
####
#### Null Hypothesis: There is no significant difference in the average charge per minute between day, evening, and night.
####
#### Alternative Hypothesis: There is a significant difference in the average charge per minute between day, evening, and night.
####


```python
import scipy.stats as stats
# Perform ANOVA test
anova_result = stats.f_oneway(df['Day Charge per Minute'],
                              df['Eve Charge per Minute'],
                              df['Night Charge per Minute'])

print(f"ANOVA F-statistic: {anova_result.statistic}")
print(f"ANOVA p-value: {anova_result.pvalue:.16f}")
```

#### Result:
#### Here the p value is less than 0.05 so we reject the null hypothesis.
####
#### Conclusion:
#### There is a significant difference in the average charge per minute between day, evening, and night.
####
#### So, the visualization difference is not by chance. Therefore, No-Churn company should focus on fixing the problem of the high charge per minute during the daytime.

```python
from sklearn.preprocessing import StandardScaler

# List of features to standardize
features_to_standerdize = [ 'Account Length', 'Day Mins', 'Eve Mins', 'Night Mins', 'International Mins', 'VMail Message', 'Day Calls', 'Eve Calls', 'Night Calls', 'International calls', 'CustServ Calls']

scaler = StandardScaler()

# Standardize the numeric features
df_new[features_to_standerdize] = scaler.fit_transform(df_new[features_to_standerdize])

# Verify the changes
df_new.head()
```
